```python
import seaborn as sns
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
print(tips.head())
```

```python
print(tips.info())
print(tips.describe())
```

```python
high_bill_tips = tips[tips['total_bill'] > 30]
print(high_bill_tips)
```

```python
sorted_tips = tips.sort_values(by='total_bill', ascending=False)
print(sorted_tips.head())
```

```python
biggest_tip = tips['tip'].max()
print(tips[tips['tip'] == biggest_tip])
```

```python
print(tips['total_bill'].corr(tips['tip']))
```

```python
#LOOPS

i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
for i in range(3):
    print("hi")
```

```python
for i in range(1, 6):
    print(i * i)
```

```python
n = 5
while n > 0:
    print(n)
    n -= 1
```

```python
for i in range(2, 10, 2):
    print(i)
```

```python
s = 0
for i in range(1, 6):
    s += i
print(s)
```

```python
for i in range(5, 0, -1):
    print(i)
```

```python
#FUNCTIONS + LIST / DICT / TUPLE

def add(a, b):
    return a + b

print(add(3, 4))
```

```python
def sq(n):
    return n * n

print(sq(5))
```

```python
def show(x):
    print(x)

show("hello")
```

```python
nums = [2, 4, 6]
def total(l):
    return sum(l)

print(total(nums))
```

```python
data = {"a": 1, "b": 2}
def get(d, k):
    return d[k]

print(get(data, "a"))
```

```python
t = (1, 2, 3)
def first(x):
    return x[0]

print(first(t))
```

```python
def even(n):
    return n % 2 == 0

print(even(8))
```

```python
names = ["A", "B", "C"]
def size(l):
    return len(l)

print(size(names))
```

```python
marks = {"x": 90, "y": 80}
def avg(d):
    return sum(d.values()) / len(d)

print(avg(marks))
```

```python
nums = (5, 10)
def mul(a, b):
    return a * b

print(mul(nums[0], nums[1]))
```

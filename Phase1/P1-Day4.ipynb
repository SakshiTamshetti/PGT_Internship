```python
# Train vs test loss

epoch = [1,2,3,4,5]
train_loss = [0.9,0.7,0.5,0.3,0.2]
test_loss = [1.0,0.8,0.6,0.5,0.4]

plt.plot(epoch, train_loss, label="Train")
plt.plot(epoch, test_loss, label="Test")
plt.legend()
plt.show()
```

```python
#Line plot with markers

plt.plot(epoch, train_loss, marker='o')
plt.show()
```

```python
# Learning rate vs accuracy

lr = [0.1, 0.01, 0.001]
accuracy = [70, 85, 90]

plt.plot(lr, accuracy)
plt.show()
```

```python
# Epoch vs loss

plt.plot(epoch, train_loss)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

# Line plot with title

plt.plot(epoch, test_loss)
plt.title("Epoch vs Test Loss")
plt.show()
```

```python
# Simple bar chart

names = ["A", "B", "C"]
values = [10, 20, 15]

plt.bar(names, values)
plt.show()

# Categorical bar chart

plt.bar(["Cat", "Dog"], [30, 50])
plt.show()

# Horizontal bar chart

plt.barh(names, values)
plt.show()

# Model accuracy comparison

models = ["SVM", "KNN", "RF"]
scores = [85, 78, 90]

plt.bar(models, scores)
plt.show()

# Stacked bar chart

train = [70, 75, 80]
test = [65, 70, 78]

plt.bar(models, train)
plt.bar(models, test, bottom=train)
plt.show()

# Bar chart with labels

plt.bar(models, scores)
for i, val in enumerate(scores):
    plt.text(i, val, val)
plt.show()

```

# Load iris dataset

```python
iris = sns.load_dataset("iris")
```

```python
plt.plot(iris["sepal_length"])
plt.title("Sepal Length Trend")
plt.show()
```

```python
plt.plot(iris["petal_length"])
plt.title("Petal Length Trend")
plt.show()
```

```python
# Sepal length vs index (species-wise)

sns.lineplot(data=iris, x=iris.index, y="sepal_length", hue="species")
plt.show()

# Petal width vs index

sns.lineplot(data=iris, x=iris.index, y="petal_width")
plt.show()

# Line plot with markers

plt.plot(iris["sepal_width"], marker="o")
plt.show()

# Line plot with grid

plt.plot(iris["petal_length"])
plt.grid()
plt.show()

# Average sepal length per species

avg_sepal = iris.groupby("species")["sepal_length"].mean()
plt.plot(avg_sepal)
plt.show()

# Average petal width per species

avg_petal = iris.groupby("species")["petal_width"].mean()
plt.plot(avg_petal)
plt.show()

```

```python
# Count of each species

iris["species"].value_counts().plot(kind="bar")
plt.show()

# Average sepal length (bar)

plt.bar(avg_sepal.index, avg_sepal.values)
plt.show()

# Average petal length

avg_len = iris.groupby("species")["petal_length"].mean()
plt.bar(avg_len.index, avg_len.values)
plt.show()

# Horizontal bar chart

plt.barh(avg_len.index, avg_len.values)
plt.show()

# Bar chart using seaborn

sns.barplot(data=iris, x="species", y="sepal_width")
plt.show()

# Species-wise petal width

sns.barplot(data=iris, x="species", y="petal_width")
plt.show()

# Bar chart with values

plt.bar(avg_petal.index, avg_petal.values)
for i, val in enumerate(avg_petal.values):
    plt.text(i, val, round(val,2))
plt.show()

# Feature comparison bar chart

features = iris.drop("species", axis=1).mean()
plt.bar(features.index, features.values)
plt.show()
```

```python
# Sepal length distribution

plt.hist(iris["sepal_length"])
plt.show()

# Petal length distribution

plt.hist(iris["petal_length"], bins=10)
plt.show()

# Sepal width histogram

plt.hist(iris["sepal_width"])
plt.show()

# Petal width histogram

plt.hist(iris["petal_width"])
plt.show()

# Histogram by species

sns.histplot(data=iris, x="sepal_length", hue="species")
plt.show()

# Density histogram

sns.histplot(iris["petal_length"], kde=True)
plt.show()

# Compare two features

plt.hist(iris["sepal_length"], alpha=0.5)
plt.hist(iris["petal_length"], alpha=0.5)
plt.show()

# Skewness detection

sns.histplot(iris["petal_width"], kde=True)
plt.show()
```

```python
# Correlation matrix

corr = iris.drop("species", axis=1).corr()
print(corr)

# Basic heatmap

sns.heatmap(corr)
plt.show()

# Annotated heatmap

sns.heatmap(corr, annot=True)
plt.show()

# Colored heatmap

sns.heatmap(corr, cmap="coolwarm")
plt.show()

# Absolute correlation heatmap

sns.heatmap(corr.abs(), annot=True)
plt.show()

# Feature selection insight

sns.heatmap(corr, linewidths=1)
plt.show()

```
